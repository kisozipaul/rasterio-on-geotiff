```python
import rasterio
import matplotlib.pyplot as plt
from rasterio.plot import show, show_hist
```

```python
img = rasterio.open('/content/drive/MyDrive/NDVI_Data/north_buganda.tif')
show(img)
```

```python
show_hist(img, bins=50)
```

```python
num_bands = img.count
print("Number of bands in the image = ", num_bands)
```

```python
img_band1 = img.read(1) #1 stands for 1st band.
img_band2 = img.read(2) #2 stands for 2nd band.
img_band3 = img.read(3) #3 stands for 3rd band.
img_band4 = img.read(4) #1 stands for 4th band.
img_band5 = img.read(5) #2 stands for 5th band.
img_band6 = img.read(6) #3 stands for 6th band.
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(img_band1, cmap='pink')
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(img_band2, cmap='pink')
ax3 = fig.add_subplot(2,2,3)
ax3.imshow(img_band3, cmap='pink')
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(img_band4, cmap='pink')
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(img_band5, cmap='pink')
ax3 = fig.add_subplot(2,2,3)
ax3.imshow(img_band3, cmap='pink')
```
